Calculating the cosine similarity

```python
from math import *
def square_root(x):
    return round(sqrt(sum([a*a for a in x])),3)

def cosine_similarity(x,y):
    numerator = sum(a*b for a,b in zip(x,y))
    deno = square_root(x) * square_root(y)
    return round((numerator/deno),3)
    
```

```python
print(cosine_similarity([3, 45, 7, 2], [2, 54, 13, 15]))
```

```python
import rake_nltk
from rake_nltk import Rake
import nltk
nltk.download('stopwords')
nltk.download('punkt')
```

```python
# initializing the new column
df['Key_words'] = ""
key_words_val = []

for index, row in df.iterrows():
    plot = row['Plot']
    
    # instantiating Rake, by default is uses english stopwords from NLTK
    # and discard all puntuation characters
    r = Rake()

    # extracting the words by passing the text
    r.extract_keywords_from_text(plot)

    # getting the dictionary whith key words and their scores
    key_words_dict_scores = r.get_word_degrees()
    
    # assigning the key words to the new column
    key_words_val.append(list(key_words_dict_scores.keys()))


```

```python
key_words_val[:5], len(key_words_val)
```

```python
df.Key_words = key_words_val
df.head()
```

```python
df.set_index('Title', inplace=True)
df.head()
```

```python
df = df.drop(columns=['Plot'])
df.head()
```

```python
for index, row in df.iterrows():
    row['Director'] = ''.join(row['Director']).lower()
    df.loc[index, 'Director'] = [row['Director']]
df.head()  
    
```

```python
columns = df.columns
df['bag_of_words'] = ''
for index, row in df.iterrows():
    words = ''
    for col in columns:
        
        words = words + ' '.join(row[col]) + ' '
    row['bag_of_words'] = words
    df.loc[index, 'bag_of_words'] = row['bag_of_words']
    
df.head()
```

```python
count = CountVectorizer()
count_matrix = count.fit_transform(df['bag_of_words'])
count_matrix
```

```python
c = count_matrix.todense()
```

```python
cos_sim = cosine_similarity(count_matrix, count_matrix)
cos_sim
```

```python
def get_recommended_movies(title, cos_sim=cos_sim):
    #print(indices)
    mov_rec = []
    idx = indices[indices==title].index[0]
    score_series = pd.Series(cos_sim[idx]).sort_values(ascending=False)
    top_10_movies_idx = score_series[:11]
    #print(top_10_movies_idx.index.values)
    for mov in top_10_movies_idx.index.values:
        #print(mov)
        mov_rec.append(indices[mov])
        
    #print(mov_rec)
    return mov_rec[1:]
```
